```python
#Вводим параметры и алгоритмы, необходимые для написания дальнейшего кода
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
def tf_idf(text):
  count = 30
  tfIdfTransformer = TfidfVectorizer(ngram_range=(1, 4), use_idf=True, max_features=count)
  countVectorizer = CountVectorizer(ngram_range=(1, 4), max_features=count)
  wordCount = countVectorizer.fit_transform([text])
  tfIdf = tfIdfTransformer.fit_transform([text])
  names = countVectorizer.get_feature_names()
  df=[]
  df = pd.DataFrame(list(names), columns=['Слова'])
  df = df.assign(WordCount=wordCount.T.todense())
  df = df.assign(TF_IDF=tfIdf.T.todense())
  df = df.sort_values('TF_IDF', ascending=False)
  print(df)
```

```python
from nltk.util import ngrams
def make_bigrams(ttexts):
    texts = ' '.join(ttexts)
    token = nltk.word_tokenize(ttexts)
    bigrams = list(ngrams(token, 2))
    # print(bigrams)
    return bigrams
def make_trigrams(ttexts):
    texts = ' '.join(ttexts)
    token = nltk.word_tokenize(ttexts)
    trigrams = list(ngrams(token, 2))
    return trigrams

btexts = make_bigrams(text) + make_trigrams(text)
id2word = corpora.Dictionary(btexts)
corpus = [id2word.doc2bow(text) for text in btexts]
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```
